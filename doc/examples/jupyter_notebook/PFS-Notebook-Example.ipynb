# Working with Versioned Pachyderm Data from Jupyter

As you run Pachyderm data pipelines the input and output of each pipeline stage or step is versioning into Pachyderm's File System (or PFS).  As certain points in the development of your pipelines, to create reports, or just to get some intuition about the data or results being generated, it is likely that you will want to access versioned data from Jupyter notebooks.  This example illustrates that process.  As you will see, after spinning up the Jupyter service, accessing PFS is a simple as accessing any other files.

## Read Data from PFS

First we are going to read in the `iris.csv` data set that we previously committed to Pachyderm data versioning system (PFS).  In a normal workflow, you might want to import data from any number of repos generated by your data pipelines to explore the data, develop new data pipeline stages, create visualizations, etc.  

Because we have specified the repo `foo` in our job spec, we are able to access the data in that repo at `/pfs/foo/`, specifically the CSV file we added `/pfs/foo/iris.csv`.

```python
cols = [
    "sepal_length",
    "sepal_width",
    "petal_length",
    "petal_width",
    "species"
]

irisDF = pd.read_csv("/pfs/foo/iris.csv", names = cols)
```

```python
irisDF.head()
```

## Explore/Visualize the Data in PFS

Now that we have read in some data from Pachyderm's file system, we can perform any of our data science tasks as we normally would in a notebook.  For example, we can visualize the data or develop additional analyses.

```python
irisDF.hist()
```

```python
irisDF.plot(kind="scatter", x="sepal_length", y="sepal_width")
```
